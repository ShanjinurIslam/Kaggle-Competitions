```python
import pandas as pd
import numpy as np
```

```python
file = open('zoo.names')

lines = file.readlines()

for line in lines:
    print(line)
```

```python
columns = ["animal_name","hair","feathers","eggs","milk","airborne","aquatic","predator","toothed","backbone","breathes","venomous","fins","legs","tail","domestic","catsize","type"]

data = pd.read_csv('zoo.data',sep=',')
```

```python
values = data.values
df = pd.DataFrame(values,columns=columns)
df.info()
```

```python
X = df.drop(columns=["animal_name","type"])
y = df.type
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

one_hot_encoder = OneHotEncoder()

transformer = ColumnTransformer([("one_hot_encoder",one_hot_encoder,X.columns)],remainder="passthrough")
```

```python
transformed_X = transformer.fit_transform(X)
y = y.astype(int)

transformed_df = pd.DataFrame(transformed_X,columns=transformer.get_feature_names())

transformed_df
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(transformed_X,y,test_size=.2,random_state=42)
```

```python
# We will test several models to find which classifies best

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
rfc.score(X_train,y_train)


print(rfc.score(X_test,y_test))
```

```python
y_pred = rfc.predict(X_test)

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score


report = classification_report(y_true=y_test,y_pred=y_pred)
print(report)
print("Accuracy : ",accuracy_score(y_test,y_pred)*100)
```

```python
# We will test several models to find which classifies best

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc.score(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

report = classification_report(y_true=y_test,y_pred=y_pred)
print(report)
print("Accuracy : ",accuracy_score(y_test,y_pred)*100)
```

```python
import os
import sklearn.tree

dotfile = open("tree.dot", "w")
sklearn.tree.export_graphviz(dtc, out_file=dotfile, feature_names=transformed_df.columns, class_names= [str(x) for x in np.unique(y)])
dotfile.close()
os.system("dot -Tpng tree.dot -o tree.png")

```
